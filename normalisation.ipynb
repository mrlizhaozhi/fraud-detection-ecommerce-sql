

## Normalise for relational database modelling

Picking dataset1 for normalisation and relational modelling. 

1. Standardise column names to remove blanks
2. Create a one-to-many relationship between Customers and Transactions tables.
   - One customer can make multiple transactions
   - One transaction can associate with only one customer
3. Subset the DataFrame
4. Further process each DataFrame to normalise
   - Remove duplicate ids
   - Remove missing ids
   - General sanity checks on data
6. Export the table files

```python
dataset1_new = dataset1.copy()
dataset1_new = dataset1_new.rename(columns={
    'Transaction ID' : 'transaction_id',
    'Customer ID' : 'customer_id',
    'Transaction Amount' : 'transaction_amount',
    'Transaction Date' : 'transaction_date',
    'Payment Method' : 'payment_method',
    'Product Category' : 'product_category',
    'Quantity' : 'quantity',
    'Customer Age' : 'age',
    'Customer Location' : 'customer_location',
    'Device Used' : 'device_used',
    'IP Address' : 'ip_address',
    'Shipping Address' : 'shipping_address',
    'Billing Address' : 'billing_address',
    'Is Fraudulent' : 'is_fraudulent',
    'Account Age Days' : 'account_age_days',
    'Transaction Hour' : 'transaction_hour'
})
```

```python
dataset1_new.head()
```

```python
customers = dataset1_new[['customer_id', 'age', 'customer_location', 'account_age_days']]
customers.head()
```

```python
customers.shape
```

```python
customers.info()
```

```python
customers = customers.drop_duplicates(subset=['customer_id'])
```

```python
transactions = dataset1_new[['transaction_id', 'customer_id', 'transaction_amount', 'transaction_date', 'payment_method', 'product_category', 'quantity', 'device_used', 'ip_address', 'shipping_address', 'billing_address', 'is_fraudulent', 'transaction_hour']]
transactions.head()
```

```python
transactions.shape
```

```python
transactions.info()
```

```python
transactions = transactions.drop_duplicates(subset=['transaction_id'])
```

## Export database tables

```python
# export the datafiles to import into pgAdmin for PostgreSQL analysis.
customers.to_csv('customers.csv', index=False, sep=",", encoding='utf-8', quotechar='"')
```

```python
transactions.to_csv('transactions.csv', index=False, sep=",", encoding='utf-8', quotechar='"')
```
